Definition of the dataset.

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
X, y = datasets.make_moons(100, noise=0.10)
x1 = X[:,0]
x2 = X[:,1]

plt.figure(figsize=(10, 7))
plt.scatter(x1, x2, c=y, cmap=plt.cm.coolwarm)
```

The dataset has 100 records with 2 features and 1 output. We need to reshape the output to make sure it has the same structure as the input features.

```python
y = y.reshape(y.shape[0], 1)

# checks the shape of the input features and output labels
print(X.shape)
print(y.shape)
```

This NN has 2 nodes in the input layer (there are 2 features in the input, X).
The hidden layer has 4 nodes.
There is 1 output layer with 1 node (because we are doing binary classification).

There are 12 weights! Their value is the difference between the predicted output and the actual output (labels).

The NN works in 2 steps: feed forward, backpropagation.

****** Feed forward step ******

```python
# creates the final output of the NN
# finds the value of zh1
# zh1 = x1w1 + x2w2 + b ---------- (1)

# finds the value of ah1
# ah1 = sigmoid(zh1) ---------- (2)

# finds remaining ah values within the hidden layer
# ah2 = sigmoid(zh2) ---------- (2)
# ah3 = sigmoid(zh3) ---------- (2)
# ah4 = sigmoid(zh4) ---------- (2)

# finds the value of zo
# zo = ah1w9 + ah2w10 + ah3w11 + ah4w12 ---------- (3)

# finds the value of NN (ao)
# ao = sigmoid(zo) ---------- (4)
```
